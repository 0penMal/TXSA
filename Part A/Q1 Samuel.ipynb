```python
import re
import nltk, string
from nltk.tokenize import *

#Function for readability
def display_columns_vertical(tokens, num_columns=4):
    col_width = 20
    rows = len(tokens) // num_columns + (1 if len(tokens) % num_columns else 0)
    
    for row in range(rows):
        line = ''
        for col in range(num_columns):
            index = row + col * rows  
            if index < len(tokens):
                line += f'{tokens[index]:<{col_width}}'
            else:
                line += ' ' * col_width
        print(line)


text = "Classification is the task of choosing the correct class label for a given input. In basic classification tasks, each input is considered in isolation from all other inputs, and the set of labels is defined in advance. The basic classification task has a number of interesting variants. For example, in multiclass classification, each instance may be assigned multiple labels; in open-class classification, the set of labels is not defined in advance; and in sequence classification, a list of inputs are jointly classified."

split_text = text.split()
token_num = len(split_text)
print(split_text)
print()
print("Total tokens:",token_num)
print()

display_columns_vertical(split_text, num_columns=4)
```

```python
#Stop tokens removal

text = "Classification is the task of choosing the correct class label for a given input. In basic classification tasks, each input is considered in isolation from all other inputs, and the set of labels is defined in advance. The basic classification task has a number of interesting variants. For example, in multiclass classification, each instance may be assigned multiple labels; in open-class classification, the set of labels is not defined in advance; and in sequence classification, a list of inputs are jointly classified."
text_lower = text.lower()
nltktokenizeforremoval_text = word_tokenize(text_lower)
token_numnltk = len(nltktokenizeforremoval_text)
print(nltktokenizeforremoval_text)
print()
print("Total tokens:",token_numnltk)

#Filtering punctuation and stop tokens
stopTokenslist = nltk.corpus.stopwords.words("english") + list(string.punctuation)
filteredTokens = []
stopTokens = []

for word in nltktokenizeforremoval_text:
    if word not in stopTokenslist:
        filteredTokens.append(word)
    else:
        stopTokens.append(word)
print()
print("There are " ,len(filteredTokens), "non stop word or punctuation tokens.")
print()
print(filteredTokens)
print()
print(stopTokens)
print("There are " ,len(stopTokens), "stop word or punctuation tokens.")
```
